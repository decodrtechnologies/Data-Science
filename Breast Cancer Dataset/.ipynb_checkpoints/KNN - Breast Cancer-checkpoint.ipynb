```python
X_train,X_test,Y_train,Y_test=train_test_split(cancer_data.data,cancer_data.target,random_state=0,test_size=0.2)
```

```python
clf=KNeighborsClassifier(n_neighbors=7)
clf.fit(X_train,Y_train)
clf.score(X_test,Y_test)
```

```python
x_axis=[]
y_axis=[]
for i in range(1,25,2):
    clf=KNeighborsClassifier(n_neighbors=i)
    x_axis.append(i)
    score=cross_val_score(clf,X_train,Y_train)
    y_axis.append(score.mean())
    print(i,score.mean())
```

```python
from collections import Counter
def predict_one(x_train,y_train,x_test,k):
    distance=[]
    for i in range(len(x_train)):
        distances=((x_train[i,:]-x_test)**2).sum()
        distance.append([distances,i])
        distance=sorted(distance)
    targets=[]
    for i in range(k):
            index_of_trainig_data=distance[i][1]
            targets.append(y_train[index_of_trainig_data])
    
    return Counter(targets).most_common(1)[0][0]    
```

```python
def predict(x_train,y_train,x_test_data,k):
    predictions=[]
    for x_test in x_test_data:
        predictions.append(predict_one(x_train,y_train,x_test,k))
    return predictions
```

```python
y_predict=predict(X_train,Y_train,X_test,7)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,y_predict)
```
